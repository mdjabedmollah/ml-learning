```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE








```

```python
data = pd.read_csv("hf://datasets/vincent-espitalier/Fashion-MNIST-CSV/fashion-mnist_train.csv")


X = data.drop('label', axis=1)
y = data['label']

print("Dataset shape:", X.shape)
```

```python

pca = PCA(n_components=50)
X_pca = pca.fit_transform(X)
print("After PCA:", X_pca.shape)
```

```python

tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X_pca)
print("After t-SNE:", X_tsne.shape)
```

```python

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=5)
plt.legend(*scatter.legend_elements(), title="Classes", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("2d Visualization")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.tight_layout()
plt.savefig("fashion pic ", dpi=300)
plt.show()

print(" Plot saved as 'fashion_mnist_pca_tsne_plot.png'")
```
