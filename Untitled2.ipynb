```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import requests
```

```python

rules_url  = "https://raw.githubusercontent.com/Sara-mibo/FPgrowth/refs/heads/main/data/rules_0.0001_0.001.csv"
data_url   = "https://raw.githubusercontent.com/mdjabedmollah/ml-learning/refs/heads/main/DataSetA.csv"
```

```python

print("Loading DataSetA (transactions)...")
raw_text = requests.get(data_url).text
lines = raw_text.strip().split("\n")

transactions = []
for line in lines:
    # remove /r
    clean = line.replace("\r", "").strip()


    try:
        items = ast.literal_eval(clean)
    except:
          # split by comma
        items = [x.strip() for x in clean.split(",") if x.strip()]

    transactions.append(items)

print("Sample transactions:")
for t in transactions[:5]:
    print(t)

print("\nTotal transactions loaded:", len(transactions))

```

```python
print("\nLoading rules dataset...")
df_rules = pd.read_csv(rules_url)
print("Rules loaded:", df_rules.shape)
```

```python
df_rules["antecedent"] = df_rules["antecedent"].apply(lambda x: ast.literal_eval(x))
df_rules["consequent"] = df_rules["consequent"].apply(lambda x: ast.literal_eval(x))

print("\nRules sample:")
print(df_rules.head())
```

```python
plt.figure(figsize=(8,4))
plt.hist(df_rules["confidence"], bins=30)
plt.title("Distribution of Rule Confidence")
plt.xlabel("Confidence")
plt.ylabel("Number of Rules")
plt.show()

plt.figure(figsize=(8,4))
plt.hist(df_rules["lift"], bins=30)
plt.title("Distribution of Rule Lift")
plt.xlabel("Lift")
plt.ylabel("Number of Rules")
plt.show()

plt.figure(figsize=(8,4))
plt.hist(df_rules["support"], bins=30)
plt.title("Distribution of Rule Support")
plt.xlabel("Support")
plt.ylabel("Number of Rules")
plt.show()
```

```python

df_rules.to_csv("clean_rules.csv", index=False)

print("\nAll tasks completed successfully! ")
```
